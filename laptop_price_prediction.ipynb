## 1. Imports & Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import shap
import pickle
import warnings

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from scipy.stats import wilcoxon

warnings.filterwarnings('ignore')
%matplotlib inline

# Reproducibility
SEED = 40
```

## 3. Data Quality Analysis

We evaluate four quality dimensions:
- **Completeness** — missing values per column
- **Validity** — values that violate known business rules
- **Consistency** — contradictions between columns (e.g. low-end CPU with dedicated GPU)
- **Duplicates** — repeated rows

```python
# --- COMPLETENESS ---
missing = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing (%)': (df.isnull().mean() * 100).round(2)
})
print("=== COMPLETENESS ===")
display(missing[missing['Missing Count'] > 0])

# Note: GPU missing (63%) is expected — many laptops have no dedicated GPU.
# Storage type missing (1.94%) is a genuine data gap.
```

```python
# --- VALIDITY ---
# Check for values that break known constraints
validity_rules = {
    'Status':       df[~df['Status'].isin({'New', 'Refurbished'})],
    'RAM <= 0':     df[df['RAM'] <= 0],
    'Storage <= 0': df[df['Storage'] <= 0],
    'Screen <= 0':  df[df['Screen'] <= 0],
    'Price <= 0':   df[df['Final Price'] <= 0],
    'Storage type': df[df['Storage type'].notna() & ~df['Storage type'].isin({'HDD', 'SSD', 'eMMC'})],
    'Touch':        df[~df['Touch'].isin({'Yes', 'No'})]
}

print("=== VALIDITY ===")
display(pd.DataFrame({
    'Rule': list(validity_rules.keys()),
    'Invalid Rows': [len(v) for v in validity_rules.values()]
}))
```

```python
# --- CONSISTENCY ---
# Business rules that should hold between columns
low_end_cpus = df['CPU'].str.contains('Celeron|Pentium', case=False, na=False)
has_gpu = df['GPU'].notna()
high_end_cpus = df['CPU'].str.contains('Core i7|Core i9|Ryzen 7|Ryzen 9', case=False, na=False)

# Detect price outliers within RAM + GPU groups
df['_gpu_flag'] = has_gpu
group_stats = df.groupby(['RAM', '_gpu_flag'])['Final Price'].agg(['mean', 'std']).reset_index()
df = df.merge(group_stats, on=['RAM', '_gpu_flag'], how='left')
price_outlier = (
    (df['Final Price'] > df['mean'] + 2 * df['std']) |
    (df['Final Price'] < df['mean'] - 2 * df['std'])
)

consistency_rules = {
    'Low-end CPU with dedicated GPU': low_end_cpus & has_gpu,
    'High-end CPU with RAM < 8GB':    high_end_cpus & (df['RAM'] < 8),
    'eMMC storage > 128GB':           (df['Storage type'] == 'eMMC') & (df['Storage'] > 128),
    'Price outlier (±2 std by group)': price_outlier
}

print("=== CONSISTENCY ===")
display(pd.DataFrame({
    'Rule': list(consistency_rules.keys()),
    'Inconsistent Rows': [int(r.sum()) for r in consistency_rules.values()]
}))

# Clean up helper columns
df = df.drop(columns=['_gpu_flag', 'mean', 'std'])
```

```python
# --- DUPLICATES ---
dup_count = df.duplicated(keep='first').sum()
print(f"Duplicate rows: {dup_count} ({dup_count/len(df)*100:.2f}%)")
# No duplicates found — each row is a unique laptop listing.
```

## 4. Data Preparation & Feature Engineering

Steps:
1. Fill missing values (GPU and Storage type → explicit 'No data' category)
2. Remove outliers (price > $6,000, screen ≤ 10", storage = 0)
3. Reduce Model cardinality (121 → top 10 + 'Other')
4. Encode categorical variables (One-Hot for nominal, Ordinal for ordered)

**Final feature matrix: 2,154 rows × 120 columns**

```python
# Reload clean and apply transformations
df = pd.read_csv('Laptops Price Dataset .csv', na_values=["NaN"])

# 1. Fill missing values with explicit category
df['Storage type'] = df['Storage type'].fillna('No data Storage')
df['GPU'] = df['GPU'].fillna('No data GPU')

# 2. Remove outliers
df = df[df['Final Price'] < 6000]
df = df[df['Screen'] > 10]
df = df[df['Storage'] > 0]

# 3. Reduce Model column cardinality (121 → top 10 + 'Other')
top10_models = df['Model'].value_counts().head(10).index.tolist()
df['Model'] = df['Model'].apply(lambda x: x if x in top10_models else 'Other')

df = df.reset_index(drop=True)
print(f"After cleaning: {df.shape}")
```

```python
# 4. Feature engineering — build final feature matrix

# Start with numeric columns
features = df[['RAM', 'Storage', 'Screen', 'Final Price']].copy()

# One-Hot Encoding for nominal categoricals
for col in ['Storage type', 'Brand', 'CPU', 'GPU', 'Model']:
    dummies = pd.get_dummies(df[col], prefix=col if col in ['Storage type'] else '')
    features = features.join(dummies)

# Ordinal Encoding for ordered categoricals
enc = OrdinalEncoder(categories=[['Refurbished', 'New']])
features['Status'] = enc.fit_transform(df[['Status']])

enc_touch = OrdinalEncoder(categories=[['No', 'Yes']])
features['Touch'] = enc_touch.fit_transform(df[['Touch']])

print(f"Feature matrix shape: {features.shape}")
features.head()
```

## 5. Feature Importance (SHAP)

We use SHAP (SHapley Additive exPlanations) to understand which features drive price predictions. Unlike standard feature importance from tree models, SHAP values show both **magnitude** and **direction** of each feature's impact.

```python
# Train a quick GBM on the 6 most interpretable features for SHAP visualization
# (Full model with 119 features is trained in section 6)
shap_features = ['RAM', 'Storage', 'Screen', 'Status', 'Touch']

# Add 'Other' model column if it exists
if 'Other' in features.columns:
    shap_features.append('Other')

X_shap = features[shap_features]
y_shap = features['Final Price']

gb_shap = GradientBoostingRegressor(random_state=SEED, max_depth=10)
gb_shap.fit(X_shap, y_shap)

explainer = shap.Explainer(gb_shap)
shap_values = explainer(X_shap)
```

```python
# Global feature importance (mean absolute SHAP value)
shap.summary_plot(shap_values, X_shap, plot_type='bar',
                  title='Global Feature Importance (mean |SHAP value|)')
```

```python
# Beeswarm plot — shows both direction and magnitude per observation
# Red = high feature value, Blue = low feature value
# Right of 0 = increases price, Left = decreases price
shap.plots.beeswarm(shap_values)
```

**Key findings:**
- **RAM** is the strongest price driver — consistently positive (more RAM → higher price)
- **Storage** follows the same pattern
- **Screen** size has a non-linear effect — very large screens don't always mean higher price
- **Status** (New vs Refurbished) creates a clear price gap

## 6. Model Selection — 375 Models Compared

We train and evaluate **375 models** across:
- **3 algorithms**: Decision Tree, Random Forest, Gradient Boosting (125 each)
- **7 feature subsets** of varying complexity
- **Multiple hyperparameter combinations**: depth, min_samples_leaf, n_estimators

Selection criterion: **lowest MAPE on validation set**

```python
# Define 7 feature subsets (from simple to full)
X_full = features.drop(columns=['Final Price'])
y_full = features['Final Price']

DATASETS = {
    'full':          features,
    'basic_numeric': features[['RAM', 'Storage', 'Screen', 'Status', 'Touch', 'Final Price']],
    'brand_top2':    features[['RAM', 'Storage', 'Screen', 'Status', '15S', 'Final Price']] if '15S' in features.columns else features[['RAM', 'Storage', 'Screen', 'Status', 'Final Price']],
    'top_brands':    features[['RAM', 'Storage', 'Screen', 'Status'] + [c for c in ['Acer','Asus','Apple','HP','Lenovo','MSI'] if c in features.columns] + ['Final Price']],
    'top_models':    features[['RAM', 'Storage', 'Screen', 'Status'] + [c for c in ['Other','15S','IdeaPad','ROG','VivoBook','ThinkPad'] if c in features.columns] + ['Final Price']],
    'top_gpus':      features[['RAM', 'Storage', 'Screen', 'Status'] + [c for c in ['No data GPU','RTX 3050','RTX 3060','RTX 3070','RTX 4060'] if c in features.columns] + ['Final Price']],
    'storage_type':  features[['RAM', 'Storage', 'Screen', 'Status'] + [c for c in ['No data Storage','eMMC','SSD'] if c in features.columns] + ['Final Price']]
}

DATAS = list(DATASETS.values())
```

```python
def evaluate_model(model, X, y):
    """Return MAE, MSE, MAPE for a fitted model on (X, y)."""
    y_pred = model.predict(X)
    return (
        mean_absolute_error(y, y_pred),
        mean_squared_error(y, y_pred),
        mean_absolute_percentage_error(y, y_pred)
    )


# Hyperparameter grids
DEPTHS_TREE    = [3, 4, 5, 6, 7, 8, 9, 10]
DEPTHS_FOREST  = [3, 5, 10, 15, None]
MIN_SAMPLES    = [2, 5, 10]
MIN_SAMPLES_TREE = [1, 2, 3, 4, 5]
N_ESTIMATORS   = [50, 100, 150, 300]
MODEL_LIMIT    = 125

TRAIN_RATIO = 0.65
VAL_RATIO   = 0.20
TEST_RATIO  = 0.15

all_results = []

for idx, df_subset in enumerate(DATAS):
    print(f"Processing dataset {idx+1}/{len(DATAS)} — {df_subset.shape[1]-1} features")

    X = df_subset.drop(columns=['Final Price'])
    y = df_subset['Final Price']

    X_train, X_val_test, y_train, y_val_test = train_test_split(
        X, y, test_size=VAL_RATIO + TEST_RATIO, random_state=SEED)
    X_val, X_test, y_val, y_test = train_test_split(
        X_val_test, y_val_test,
        test_size=TEST_RATIO / (TEST_RATIO + VAL_RATIO),
        random_state=SEED)

    k_list_tree   = [min(k, X_train.shape[1]) for k in [3, 5, 7, 9, X_train.shape[1]]]
    k_list_forest = [min(k, X_train.shape[1]) for k in [5, 10, X_train.shape[1]]]

    results = []
    counters = {'tree': 0, 'forest': 0, 'boost': 0}

    for model_type in ['tree', 'forest', 'boost']:
        depths = DEPTHS_TREE if model_type == 'tree' else DEPTHS_FOREST
        min_s  = MIN_SAMPLES_TREE if model_type == 'tree' else MIN_SAMPLES
        k_list = k_list_tree if model_type == 'tree' else k_list_forest
        n_est  = [None] if model_type == 'tree' else N_ESTIMATORS

        for depth in depths:
            for ms in min_s:
                for k in k_list:
                    selector = SelectKBest(score_func=f_regression, k=k)
                    Xtr = selector.fit_transform(X_train, y_train)
                    Xv  = selector.transform(X_val)
                    Xte = selector.transform(X_test)

                    for ne in n_est:
                        if counters[model_type] >= MODEL_LIMIT:
                            break

                        if model_type == 'tree':
                            m = DecisionTreeRegressor(max_depth=depth, min_samples_leaf=ms, random_state=SEED)
                        elif model_type == 'forest':
                            m = RandomForestRegressor(n_estimators=ne, max_depth=depth, min_samples_leaf=ms, random_state=SEED)
                        else:
                            m = GradientBoostingRegressor(n_estimators=ne, max_depth=depth, min_samples_leaf=ms, random_state=SEED)

                        m.fit(Xtr, y_train)
                        mae_v, mse_v, mape_v = evaluate_model(m, Xv, y_val)

                        results.append({
                            'model_type': model_type, 'max_depth': depth,
                            'min_samples_leaf': ms, 'n_estimators': ne, 'k_features': k,
                            'mae_val': mae_v, 'mse_val': mse_v, 'mape_val': mape_v,
                            'model': m, 'x_test_sel': Xte, 'y_test': y_test
                        })
                        counters[model_type] += 1

                    if counters[model_type] >= MODEL_LIMIT: break
                if counters[model_type] >= MODEL_LIMIT: break
            if counters[model_type] >= MODEL_LIMIT: break

    results_df = pd.DataFrame(results)
    best = results_df.sort_values('mape_val').iloc[0]
    mae_t, mse_t, mape_t = evaluate_model(best['model'], best['x_test_sel'], best['y_test'])

    all_results.append({
        'dataset': idx, 'model_type': best['model_type'],
        'max_depth': best['max_depth'], 'min_samples_leaf': best['min_samples_leaf'],
        'n_estimators': best['n_estimators'], 'k_features': best['k_features'],
        'MAE_val': best['mae_val'], 'MSE_val': best['mse_val'], 'MAPE_val': best['mape_val'],
        'MAE_test': mae_t, 'MSE_test': mse_t, 'MAPE_test': mape_t,
        'model': best['model']
    })

final_results_df = pd.DataFrame(all_results)
print("\n=== Best model per dataset ===")
display(final_results_df.drop(columns=['model']))
```

## 7. Final Model Evaluation

We select the overall best model and evaluate it thoroughly:
- Point metrics (MAE, MSE, MAPE)
- 95% confidence intervals via bootstrap
- Error by brand
- Top-10 most important features
- Statistical significance test vs. mean-price baseline (Wilcoxon)

```python
# Train the final model on the full dataset (full feature set)
X_all = features.drop(columns=['Final Price'])
y_all = features['Final Price']

X_train, X_val_test, y_train, y_val_test = train_test_split(
    X_all, y_all, test_size=VAL_RATIO + TEST_RATIO, random_state=SEED)
X_val, X_test, y_val, y_test = train_test_split(
    X_val_test, y_val_test,
    test_size=TEST_RATIO / (TEST_RATIO + VAL_RATIO),
    random_state=SEED)

# Best configuration found during model selection
final_model = GradientBoostingRegressor(
    n_estimators=300, max_depth=5, min_samples_leaf=2, random_state=SEED
)

# Train on train + validation, evaluate on test
X_train_val = pd.concat([X_train, X_val])
y_train_val = pd.concat([y_train, y_val])
final_model.fit(X_train_val, y_train_val)

y_pred = final_model.predict(X_test)
```

```python
# Point metrics
mae  = mean_absolute_error(y_test, y_pred)
mse  = mean_squared_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print("=== Final Model — Test Set Metrics ===")
print(f"  MAE:  ${mae:.2f}")
print(f"  MSE:  {mse:.2f}")
print(f"  MAPE: {mape*100:.2f}%")
print(f"\n  Average laptop price: ${y_test.mean():.0f}")
print(f"  → Model is off by ~${mae:.0f} on average ({mape*100:.1f}% of price)")
```

```python
# 95% Confidence intervals for each metric
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    margin = stats.sem(data) * stats.t.ppf((1 + confidence) / 2., n - 1)
    return mean, mean - margin, mean + margin

residuals = y_pred - y_test.values
abs_errors = np.abs(residuals)
sq_errors  = residuals ** 2
pct_errors = abs_errors / y_test.values * 100

mae_m,  mae_lo,  mae_hi  = confidence_interval(abs_errors)
mse_m,  mse_lo,  mse_hi  = confidence_interval(sq_errors)
mape_m, mape_lo, mape_hi = confidence_interval(pct_errors)

print("=== 95% Confidence Intervals ===")
print(f"  MAE:  (${mae_lo:.2f}, ${mae_hi:.2f})")
print(f"  MSE:  ({mse_lo:.2f}, {mse_hi:.2f})")
print(f"  MAPE: ({mape_lo:.2f}%, {mape_hi:.2f}%)")
```

```python
# Error by brand — which brands are harder to predict?
test_with_brand = df.loc[X_test.index, ['Brand']].copy()
test_with_brand['abs_error'] = abs_errors

brand_errors = (
    test_with_brand.groupby('Brand')['abs_error']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 5))
plt.bar(brand_errors.index, brand_errors.values, color='steelblue')
plt.title('Top 10 Brands by Mean Absolute Error')
plt.ylabel('Mean Absolute Error ($)')
plt.xlabel('Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 most important features by the final model
importances = pd.Series(
    final_model.feature_importances_,
    index=X_train_val.columns
).sort_values(ascending=False).head(10)

importances.plot.bar(figsize=(10, 5), color='steelblue')
plt.title('Top 10 Features — Model Importance')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Wilcoxon test — is our model significantly better than a naive mean-price baseline?
mean_baseline = np.full_like(y_test.values, fill_value=y_train_val.mean())
mape_model    = np.abs((y_test.values - y_pred) / y_test.values) * 100
mape_baseline = np.abs((y_test.values - mean_baseline) / y_test.values) * 100

stat, p_value = wilcoxon(mape_model, mape_baseline, alternative='less')

print("=== Wilcoxon Test: Model vs. Mean Baseline ===")
print(f"  Statistic: {stat}")
print(f"  p-value:   {p_value:.2e}")
print(f"  → {'Model is significantly better (p < 0.001)' if p_value < 0.001 else 'No significant difference'}")
```
